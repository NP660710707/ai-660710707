##เทคนิคการจัดการข้อมูลพื้นฐาน

 *การจัดการข้อมูลหาย*

ลบแถวที่มีค่าหาย

```python
# เหมาะเมื่อมีข้อมูลหายน้อย - ลบแถวที่มี NaN (ค่าหาย)
df_clean = df.dropna()
```

```python
# DataFrame ใหม่ที่ลบ NaN (ค่าหาย) ออกแล้ว
df_clean
```

เติมค่าด้วยค่าเฉลี่ย

```python
#หาค่าเฉลี่ยของคอลัมน์อายุ
mean_age = df['อายุ'].mean()
```

```python
mean_age
```

```python
#เติมค่า 28.470588 ลงในช่องที่เป็น NaN (ของ 64003, ธีระ, รัตนา)
df['อายุ'] = df['อายุ'].fillna(mean_age)
```

```python
df
```

เติมค่าด้วยค่าที่พบบ่อยที่สุด

```python
#หาค่าที่พบบ่อยที่สุด (Mode) ในคอลัมน์คณะ ในนี้จะเป็น Science และจะเติม 'Science' ลงในช่องที่เป็น NaN (ของธีระ, ยุพา)
df['คณะ'].fillna(df['คณะ'].mode()[0], inplace=True)
```

```python
df
```

การทำให้ข้อมูลเป็นมาตรฐาน

```python
#ทำให้ชื่อคณะเป็นมาตรฐานเดียวกัน
def standardize_faculty(name):
    if name in ['วิทยาศาสตร์', 'วิทย์', 'Science', 'คณะวิทยาศาสตร์', 'Sci']:
        return 'วิทยาศาสตร์'
    elif name in ['ศิลปศาสตร์', 'ศิลป์', 'Arts', 'Liberal Arts']:
        return 'ศิลปศาสตร์'
    elif name in ['วิศวฯ', 'วิศวะ', 'วิศวกรรม', 'Engineering']:
        return 'วิศวกรรมศาสตร์'
    else:
        return name
```

```python
#สร้างคอลัมน์ใหม่ที่เก็บชื่อคณะที่เป็นมาตรฐานแล้ว
df['คณะ_มาตรฐาน'] = df['คณะ'].apply(standardize_faculty)
```

```python
df['คณะ_มาตรฐาน']
```

การจัดการค่าผิดปกติ

```python
print(f"\nค่าเฉลี่ยอายุ (รวมค่าผิดปกติ) - {df['อายุ'].mean():.2f}")
```

```python
#หาค่า Quartile ที่ 1 และ หาค่า Quartile ที่ 3
Q1 = df['อายุ'].quantile(0.25)
Q3 = df['อายุ'].quantile(0.75)
```

```python
Q1, Q3
```

```python
#Interquartile Range คือช่วงระหว่าง Q1 และ Q3 บอกถึงการกระจายตัวของข้อมูลส่วนกลาง 50%
IQR = Q3 - Q1
```

```python
IQR
```

```python
#กำหนดขอบเขต
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
lower_bound, upper_bound
```

```python
#กรองค่าผิดปกติ
df_no_outliers = df[(df['อายุ'] >= lower_bound) & (df['อายุ'] <= upper_bound)]
```

```python
df_no_outliers
```

```python
#กรองเฉพาะแถวที่มีอายุอยู่ในช่วงที่กำหนด
(df['อายุ'] >= lower_bound) & (df['อายุ'] <= upper_bound)
```

##Feature Engineering

One-Hot Encoding แปลงข้อมูลประเภทเป็นตัวเลข

```python
data = pd.DataFrame(student)
```

```python
#แปลงคอลัมน์คณะเป็น  One-Hot Encoding
data_encoded = pd.get_dummies(data, columns=['คณะ'])
print(data_encoded)
```

การสร้าง Feature จากวันเวลา

```python
data['วันที่ลงทะเบียน'] = pd.to_datetime(data['วันที่ลงทะเบียน'])
data['วันในสัปดาห์'] = data['วันที่ลงทะเบียน'].dt.day_name()
data
```

การรวม Features

```python
#สร้าง BMI
data['BMI'] = data['น้ำหนัก_kg'] / (data['ส่วนสูง_cm']/100)**2
```

```python
data
```

```python
data['สุขภาพ'] = pd.cut(data['BMI'],
                                   bins=[0, 18.5, 25, 30, 100],
                                   labels=['ผอม', 'ปกติ', 'อ้วน', 'อ้วนมาก'])
```
